```python
# Import Modules
import pandas as pd
import numpy as np
import re
import nltk
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('recommendation.csv')
df.head()
```

```python
len(df)
```

```python
df['Plot'][0]
```

```python
# Convert the plot to lowercase
df['cleanPlot'] = df['Plot'].str.lower()
df['cleanPlot'] = df['cleanPlot'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))
df['cleanPlot'] = df['cleanPlot'].apply(lambda x: re.sub('\s+', ' ', x))
df['cleanPlot']
```

```python
# Tokenization
nltk.download('punkt')
df['cleanPlot'] = df['cleanPlot'].apply(lambda x: nltk.word_tokenize(x))
df['cleanPlot']
```
